```python
# Import the necessary libraries
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()

# Load the data
tips = sns.load_dataset("tips")

# Create violinplot
sns.violinplot(x = "total_bill", y="sex",data=tips)

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
td = np.arange(0,10,.5)
td.resize(td.shape[0],1)
```

```python
temp = td**2 + np.random.normal(0,.5,td.shape)
```

```python
temp.shape
```

```python
td = np.concatenate([td,temp],axis = 1)
```

```python
import sklearn.linear_model as lm
```

```python
lreg = lm.LinearRegression()
```

```python
lreg.fit(td[:,0:1]**2,td[:,1])
```

```python
xs = np.array([[0],[100]])
ys = lreg.predict(xs)
```

```python
sns.lineplot(x = xs[:,0], y = ys) + sns.stripplot(x = td[:,0]**2,y = td[:,1])
```
